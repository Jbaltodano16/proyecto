```python
import numpy as np

# 1. Convertir a mayúsculas antes de reemplazar (mejor para asegurar coincidencias)
dfpropiedades['property_type'] = dfpropiedades['property_type'].str.upper()

# 2. Reemplazar el texto basura por NaN
dfpropiedades['property_type'] = dfpropiedades['property_type'].replace(
    ';;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;;',
    np.nan
)

# 3. Verificar si 'COMMERCIAL' está presente antes de rellenar NaN
if 'COMMERCIAL' not in dfpropiedades['property_type'].unique():
    # Si 'COMMERCIAL' no está, rellenar los NaN con la categoría más frecuente
    valor_mas_comun = dfpropiedades['property_type'].mode()[0]
    dfpropiedades['property_type'] = dfpropiedades['property_type'].fillna(valor_mas_comun)
else:
    # Si 'COMMERCIAL' está presente, no cambiar NaN por la moda
    dfpropiedades['property_type'] = dfpropiedades['property_type'].fillna('COMMERCIAL')

# 4. Eliminar duplicados
dfpropiedades.drop_duplicates(inplace=True)

# 5. Eliminar otras filas con NaN si quedan
dfpropiedades.dropna(inplace=True)

# 6. Verificar el resultado
print("DataFrame limpio")
print(dfpropiedades.head())
```

```python
def agrupaYcuenta(dfParam, nomColumna, orden): #gf es un df para graficar
    gf=dfParam.groupby([nomColumna]).count()
    gf=gf.iloc[:,:1] #permite filtrar y seleccionar columnas y observaciones, al colocarlo asi [:] él entiende que se quieren todas las observaciones
    gf.rename({"serial_number":"cantidad"}, axis=1, inplace=True)
    gf=gf.sort_values('cantidad',ascending=orden)
    return gf
```

```python
# Assuming dfAgrupado is already defined
dfAgrupado = agrupaYcuenta(dfpropiedades, 'sale_amount', False)  # <- reemplazá con tus valores reales
dfAgrupado = dfAgrupado.head(5)

fig, ax = plt.subplots(1, 1, figsize=(15, 6.9), dpi=100)

etiquetasBarras = dfAgrupado.index.tolist()
valores = dfAgrupado['cantidad'].tolist()

ypos = np.arange(len(etiquetasBarras))
colores = ["#a2caa5", "#e5eaa4", "#e5eaa4", "#e5eaa4", "#eb7072"]

barras = ax.barh(ypos, valores, color=colores, align='center', alpha=0.8, edgecolor='black')
ax.set_yticks(ypos, labels=etiquetasBarras)

titulo = "Grafico 1: Propiedades mas vendidas"
plt.suptitle(titulo,
             horizontalalignment='left',
             x=0.001,
             y=0.87,
             transform=fig.transFigure,
             fontsize=20,
             fontweight='bold',
             fontfamily='serif')

subtitulo = "Ingreso anual de las propiedades mas vendidas"
ax.set_title(subtitulo,
             ha='left',
             x=0.001,
             y=0.78,
             transform=fig.transFigure,
             fontsize=12,
             fontweight='regular',
             fontfamily='serif')

etiquetax = "Cantidades de propiedades"
ax.set_xlabel(etiquetax,
              ha='left',
              fontsize=10,
              fontweight='regular',
              fontfamily='serif')

ax.xaxis.set_label_coords(x=0, y=-0.07, transform=ax.transAxes)

for barra in barras:
    width = barra.get_width()
    ax.text(width + 0.20, barra.get_y() + barra.get_height() / 2, f'{width:.0f}', ha='left', va='center', fontsize=10, color='black')

plt.subplots_adjust(top=0.75)
plt.show()
```

```python

# Asegurar que los valores de 'assesed_value' sean numéricos
dfpropiedades['assesed_value'] = pd.to_numeric(dfpropiedades['assesed_value'], errors='coerce')

# Calcular promedios
residential_avg = dfpropiedades[dfpropiedades['property_type'] == 'RESIDENTIAL']['assesed_value'].mean()
commercial_avg = dfpropiedades[dfpropiedades['property_type'] == 'COMMERCIAL']['assesed_value'].mean()

# Ajustar valor mínimo
commercial_avg = commercial_avg if commercial_avg > 0 else 1

# Crear DataFrame para graficar
data_df = pd.DataFrame({
    'Tipo de Propiedad': ['RESIDENTIAL', 'COMMERCIAL'],
    'Costo Promedio': [residential_avg, commercial_avg]
})

# Crear gráfico
fig, ax = plt.subplots(figsize=(8, 5), dpi=100)
ax.bar(data_df['Tipo de Propiedad'], data_df['Costo Promedio'], color=['#a2caa5', '#f4a460'])

# Configuración del eje Y
ax.set_ylim(0, data_df['Costo Promedio'].max() * 1.1)

# Etiquetas
ax.set_xlabel("Tipo de Propiedad", fontsize=12, fontweight="bold")
ax.set_ylabel("Costo Promedio", fontsize=12, fontweight="bold")
ax.set_title("Grafica #4 Costo Promedio entre Propiedades Residenciales y Comerciales", fontsize=16, fontweight="bold")
# Añadir valores a las barras
for i, valor in enumerate(data_df['Costo Promedio']):
    ax.text(i, valor + 1, f'{valor:,.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

<h4>Pregunta de Negocio #6</h4>
<h5>¿Cuál es el margen promedio de ganancia por cada una de las propiedades vendidas?</h5>

```python
# Convertir columnas a numéricas
dfpropiedades['assesed_value'] = pd.to_numeric(dfpropiedades['assesed_value'], errors='coerce')
dfpropiedades['sale_amount'] = pd.to_numeric(dfpropiedades['sale_amount'], errors='coerce')

# Reemplazar NaN con el promedio de la columna
dfpropiedades['assesed_value'].fillna(dfpropiedades['assesed_value'].mean(), inplace=True)
dfpropiedades['sale_amount'].fillna(dfpropiedades['sale_amount'].mean(), inplace=True)

# Calcular margen de ganancia
dfpropiedades['margen_ganancia'] = (dfpropiedades['sale_amount'] - dfpropiedades['assesed_value']) / dfpropiedades['assesed_value']
```

```python

dfpropiedades['sale_amount'] = pd.to_numeric(dfpropiedades['sale_amount'], errors='coerce')
dfpropiedades['assesed_value'] = pd.to_numeric(dfpropiedades['assesed_value'], errors='coerce')

dfpropiedades['sale_amount'].fillna(dfpropiedades['sale_amount'].mean(), inplace=True)
dfpropiedades['assesed_value'].fillna(dfpropiedades['assesed_value'].mean(), inplace=True)


dfpropiedades['margen_pct'] = ((dfpropiedades['sale_amount'] - dfpropiedades['assesed_value'])
                               / dfpropiedades['assesed_value']) * 100
dfpropiedades['margen_pct'].replace([np.inf, -np.inf], np.nan, inplace=True)


tipos_res = ['Single Family','Two Family','Three Family','Four Family','Condo']
df_resid = dfpropiedades[dfpropiedades['residential_type'].isin(tipos_res)]
df_resid = df_resid.dropna(subset=['margen_pct'])


margen_prom = df_resid.groupby('residential_type')['margen_pct'].mean()
if margen_prom.empty:
    print("No hay datos válidos para graficar.")
else:
    margen_prom = margen_prom.sort_values(ascending=False)

    fig, ax = plt.subplots(figsize=(10,6), dpi=100)
    barras = margen_prom.plot(kind='bar', ax=ax,
                              color=['#6f0550','#fad889','#8d99ae','#ff5e6c','#7cc6fe'])

    ax.set_ylabel("Margen Promedio (%)", fontsize=12, fontweight='bold')
    ax.set_xlabel("Tipo de Propiedad", fontsize=12, fontweight='bold')
    ax.set_title("Margen Promedio de Ganancia por Tipo Residencial", fontsize=14, fontweight='bold')

    # Ajustar límites solo si son números
    ymax = margen_prom.max() * 1.1
    ax.set_ylim(0, ymax if np.isfinite(ymax) else 0)

    ax.set_xticklabels(margen_prom.index, rotation=0)

    # Etiquetas de valor
    for i, val in enumerate(margen_prom):
        ax.text(i, val + 1, f"{val:.1f}%", ha='center', va='bottom', fontsize=10)

    plt.tight_layout()
    plt.show()
```
